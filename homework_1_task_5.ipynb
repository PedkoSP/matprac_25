```python
class Test:
    @staticmethod
    def assert_equals(a, b, *args, **kwargs):
        assert a == b
        print('Passed')
```

```python
from random import randint

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def add_data(self, lst):
        current = self
        while lst:
            current.next = ListNode(lst[0])
            current = current.next
            lst = lst[1:]

    def get_data(self):
        res = []
        current = self
        while current:
            res.append(current.val)
            current = current.next
        return res

def add_two_numbers(x, y):
    lx = x.get_data()
    ly = y.get_data()

    lres = []
    i = 0
    add = 0
    
    while True:
        if (i >= len(lx)) & (i >= len(ly)):
            if add == 1:
                lres.append(1)
            res = ListNode(lres[0])
            res.add_data(lres[1:])
            return res
        if (i >= len(lx)) & (i < len(ly)):
            r = (ly[i] + add) % 10
            if ly[i] + add > 9:
                add = 1
            else:
                add = 0
            lres.append(r)
        if (i < len(lx)) & (i >= len(ly)):
            r = (lx[i] + add) % 10
            if lx[i] + add > 9:
                add = 1
            else:
                add = 0
            lres.append(r)
        if (i < len(lx)) & (i < len(ly)):
            r = (lx[i] + ly[i] + add) % 10
            if lx[i] + ly[i] + add > 9:
                add = 1
            else:
                add = 0
            lres.append(r)
        i += 1
```

```python
from time import perf_counter
tic = perf_counter()

lt1 = ListNode(2)
lt1.add_data([4, 3])
lt2 = ListNode(5)
lt2.add_data([6, 4])
Test.assert_equals(add_two_numbers(lt1, lt2).get_data(), [7, 0, 8])

lt1 = ListNode(0)
lt2 = ListNode(0)
Test.assert_equals(add_two_numbers(lt1, lt2).get_data(), [0])

lt1 = ListNode(9)
lt1.add_data([9,9,9,9,9,9])
lt2 = ListNode(9)
lt2.add_data([9,9,9])
Test.assert_equals(add_two_numbers(lt1, lt2).get_data(), [8, 9, 9, 9, 0, 0, 0, 1])

n_max = pow(10, 20)
for _ in range(100):
    n1 = randint(0, n_max)
    n2 = randint(0, n_max)
    expected = list(map(int, list(str(n1 + n2))))[::-1]
    lst1 = list(map(int, list(str(n1))))[::-1]
    lst2 = list(map(int, list(str(n2))))[::-1]
    ll1 = ListNode(lst1[0])
    ll1.add_data(lst1[1:])
    ll2 = ListNode(lst2[0])
    ll2.add_data(lst2[1:])
    Test.assert_equals(add_two_numbers(ll1, ll2).get_data(), expected)

print('t_sec = {:.6f}'.format(perf_counter() - tic))
```
